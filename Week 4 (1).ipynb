## 2) Data

The data needed for this project comes from 3 different sources. The first source is a Wikepedia listing the postal codes of Canada, specifically the Toronto area. This data consists of the Postal Code, the Borough name, and the Neighborhood name.

The next data source is geospatial data that is used for longitude and latitude coordinates, which is retreieved from a csv file.

The last data source used for this project comes from the FourSquare API and gathers Neighborhood name, Latitude, Longitude, Venue name, Venue Latitude, Venue Longitude, and Venue Category.
