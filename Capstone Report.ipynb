### Introduction

This project will analyze geographic data of the Scarborough borough in Toronto, Canada. A small business owner is looking to open a store in the city and is deciding between opening a coffee shop and a gas station. Additionally, they need to identify the best neighborhood to open the store in order to avoid competition and be as profitable as possible. To acheive this, this project will explore the neighborhoods as well as the venues that exist in them, aiming to find the best-fitting neighborhood for the new coffee shop/gas station.

### Data

The data needed for this project comes from 3 different sources. The first source is a Wikepedia listing the postal codes of Canada, specifically the Toronto area. This data consists of the Postal Code, the Borough name, and the Neighborhood name. The next data source is geospatial data that is used for longitude and latitude coordinates, which is retreieved from a csv file. The last data source used for this project comes from the FourSquare API and gathers Neighborhood name, Latitude, Longitude, Venue name, Venue Latitude, Venue Longitude, and Venue Category.

### Methodology

##### Work Flow:

1. Data for neighborhoods in Toronto is gathered from the Wikepedia page and is pre-processed before ultimately joining with the coordinates csv file. This neighborhod data is then filtered down to only contain Scarborough neigborhoods.
2. The venue data is fetched from the Foursquare API. Each venue can then be assigned to a neighborhood and joined with the neighborhood data from Wikipedia. 
3. The venue data is grouped together by venue category, allowing for a good summary of the venues that exist within each neighborhood. Conclusions can be derived from these findings to provide business value for the client. 

### Results

Looking at the Scarbourough borough, there are 15 neighborhoods that were analyzed. In order to decide whether our client should pursue opening a coffee shop or a gas station, we looked at the total count for each of them in the entire borough. Competition can exist across neighborhood lines, so we are looking for the venue that is the most rare across the entire borough.

### Discussion

The results reveal that we would reccommend opening a gas station, as there are less throughout the entire borough. 


Additionally, we would recommend that the gas station be located in either Golden Mile, Guildwood, Wexfor, or Woburn. These neighborhoods do not feature a gas station as one oft their top 10 most frequent venues, meaning there is not a high level of competition for business within these neighborhoods. 

To boil it down to a single neighborhood recommendation, we would select Guildwood, as they have a small number of grocery stores and a fair amount of intersections, which are both good for attracting customers to a gas station.
